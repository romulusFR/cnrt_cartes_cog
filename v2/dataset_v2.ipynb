```python
import pandas as pd
from collections import Counter
from IPython.display import display
```

```python
df = pd.read_excel("BDD_2022.xlsx")
```

```python
mine_cols = [f"Mine {i}" for i in range(1, 20)]
df_mine = df[mine_cols].astype("string")
display(df_mine)

```

```python
dict_mine = {idx:[val.lower().strip() for val in vals.dropna().tolist()] for idx, vals in df_mine.iterrows()}
```

```python
all_words = Counter(w for l in dict_mine.values() for w in l)
display(all_words.most_common(20))
```

```python
def get_n_words(n:int):
    return [vals[n] for vals in dict_mine.values() if n < len(vals) ]

for i in range(20):
    n_words = get_n_words(i)
    print(f"{len(n_words)} in pos {i}")
```
